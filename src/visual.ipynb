```python
import json
from matplotlib import pyplot as plt

with open("trained-results.json", "r") as f:
    train_data = json.load(f)

with open("base-results.json", "r") as f:
    baseline_data = json.load(f)
```

```python
print(len(train_data["rouge1"]))
print(len(baseline_data["rouge1"]))
```

```python
plt.hist(train_data["rouge1"], label="train", alpha=0.5, bins=20)
plt.hist(baseline_data["rouge1"], label="baseline", alpha=0.5, bins=20)
plt.legend()
plt.show()
```

```python
from statistics import mean, median

for key in train_data.keys():
    print(key)
    print(f"train: {round(mean(train_data[key]), 2)}, baseline: {round(mean(baseline_data[key]), 2)}")
```

```python
plt.hist(train_data["rougeLsum"], label="train", alpha=0.5, bins=20)
plt.hist(baseline_data["rougeLsum"], label="baseline", alpha=0.5, bins=20)
plt.legend()
plt.show()
```
